# Calculating the mean of close value for certain time windows

# Price change compare to previous day and final closing price for the day is correlated or not

```python
closing_price.head()
```

```python
closing_price['apple_close']
```

```python
closing_price['apple_close'].shift(1)
```

```python
#pct of changing price
(closing_price['apple_close']-closing_price['apple_close'].shift(1))/closing_price['apple_close'].shift(1)*100
```

```python
for col in closing_price.columns:
    closing_price[col+'pct_change']= (closing_price[col]-closing_price[col].shift(1))/closing_price[col].shift(1)*100
```

```python
closing_price
```

```python
closing_price.columns
```

```python
closing_pp= closing_price[[ 'apple_closepct_change', 'amazon_closepct_change',
       'google_closepct_change', 'microsoft_closepct_change']]
```

```python
closing_pp
```

```python
sns.pairplot(closing_pp)
```

```python
v = sns.PairGrid(closing_pp)

v.map_diag(sns.histplot)
v.map_lower(sns.scatterplot)
v.map_upper(sns.kdeplot)
```
